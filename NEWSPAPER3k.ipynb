# Download Entire CNN Paper

```python
import newspaper

cnn_paper = newspaper.build('http://cnn.com')
cnn_article = cnn_paper.articles[0]
cnn_article.download()
cnn_article.parse()
cnn_article.nlp()
```

```python
for article in cnn_paper.articles:
    print(article.url)
```

```python
for category in cnn_paper.category_urls():
    print(category)
```
